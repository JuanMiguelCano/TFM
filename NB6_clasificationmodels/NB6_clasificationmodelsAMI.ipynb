```r
library(bnlearn)
library(caret)
library(randomForest)
library(pROC)
library(e1071)  # for SVM and Naive Bayes
library(class)  # for KNN
library(ROSE)   # for undersampling
library(ggplot2) # for plots
```

```r
# Load data
data <- read.csv("DATA.csv")

# Filter the data into predictors and target variable
predictors <- grep("^MH|^TRT|^SOC", colnames(data), value = TRUE)
variables <- c(predictors, "AMI")
data_subset <- data[, variables] # Subset the data with the selected variables

# Convert to factors
for (col in colnames(data_subset)) {
  if (is.integer(data_subset[[col]])) {
    data_subset[[col]] <- as.factor(data_subset[[col]])
  }
}
data_subset$AMI <- as.factor(data_subset$AMI)

# Function to perform undersampling
undersample_data <- function(data) {
  data_balanced <- ovun.sample(AMI ~ ., data = data, method = "under")$data
  return(data_balanced)
}

match_levels <- function(train_data, test_data){ # Function to ensure categorical levels match
  for (col in colnames(train_data)) {
    if (is.factor(train_data[[col]])) {
      levels(test_data[[col]]) <- levels(train_data[[col]])
    }
  }
  return(test_data)
}

# Function to train and evaluate the model
train_and_evaluate <- function(train_data, test_data, model_type) {
  test_data <- match_levels(train_data, test_data)
  if (model_type == "bn"){
    structure <- hc(train_data) # learn the structure of the Bayesian Network with Hill-Climbing
    bn_model <- bn.fit(structure, train_data)
    predictions <- sapply(1:nrow(test_data), function(i){ # Make predictions using cpquery
      evidence <- test_data[i, -which(names(test_data) == "AMI")]
      if (is.null(evidence)) return(NA)
      cpquery(bn_model, event = (AMI == "1"), evidence = as.list(evidence), method = "lw")
    })
    pred_class <- factor(ifelse(predictions > 0.5, "1", "0"), levels = c("0", "1")) # Convert predictions to classes
  } else if (model_type == "rf"){
    rf_model <- randomForest(AMI ~ ., data = train_data, ntree = 100, importance = TRUE)
    pred_class <- predict(rf_model, test_data)
  } else if (model_type == "svm"){
    svm_model <- svm(AMI ~ ., data = train_data, kernel = "radial", probability = TRUE)
    predictions <- predict(svm_model, newdata = test_data, probability = TRUE)
    pred_prob <- attr(predictions, "probabilities")[, 2]
    pred_class <- factor(ifelse(pred_prob > 0.5, "1", "0"), levels = c("0", "1"))
  } else if (model_type == "knn"){
    pred_class <- knn(train = train_data[, -which(names(train_data) == "AMI")], 
                      test = test_data[, -which(names(test_data) == "AMI")], 
                      cl = train_data$AMI, 
                      k = 5)
  } else if (model_type == "nb") {
    nb_model <- naiveBayes(AMI ~ ., data = train_data)
    predictions <- predict(nb_model, newdata = test_data, type = "raw")[, 2]
    pred_class <- factor(ifelse(predictions > 0.5, "1", "0"), levels = c("0", "1"))
  } else {
    stop("Invalid. Supported models: bn, rf, svm, knn, nb")
  }
  return(list(pred_class = pred_class, real = test_data$AMI)) # Return predictions and the actual class
}

num_iterations <- 10 # Number of iterations
auc_results <- data.frame( # data frame to store AUC results
  AUC = numeric(),
  Model = factor(),
  Iteration = integer()
)

# Initialize lists to store specificity and sensitivity for each model
specificity_list <- list(
  "Bayesian Network" = list(),
  "Random Forest" = list(),
  "SVM" = list(),
  "KNN" = list(),
  "Naive Bayes" = list()
)
sensitivity_list <- list(
  "Bayesian Network" = list(),
  "Random Forest" = list(),
  "SVM" = list(),
  "KNN" = list(),
  "Naive Bayes" = list()
)

for (iter in 1:num_iterations) {
  cat("\nIteration",iter,"\n")
  set.seed(iter) # Perform undersampling with a different seed for each iteration
  data_subset_balanced <- undersample_data(data_subset)
  # Perform LOOCV on the balanced dataset
  n <- nrow(data_subset_balanced)
  # Initialize vectors to store results
  pred_bn <- factor(levels = c("0", "1"))
  pred_rf <- factor(levels = c("0", "1"))
  pred_svm <- factor(levels = c("0", "1"))
  pred_knn <- factor(levels = c("0", "1"))
  pred_nb <- factor(levels = c("0", "1"))
  real_classes <- factor(levels = c("0", "1"))

  for (i in 1:n) {
    train_data <- data_subset_balanced[-i,]
    test_data <- data_subset_balanced[i, , drop = FALSE]
    result_bn <- train_and_evaluate(train_data, test_data, "bn") # Evaluate Bayesian Network
    pred_bn <- c(pred_bn, result_bn$pred_class)
    result_rf <- train_and_evaluate(train_data, test_data, "rf") # Evaluate Random Forest
    pred_rf <- c(pred_rf, result_rf$pred_class)
    result_svm <- train_and_evaluate(train_data, test_data, "svm") # Evaluate SVM
    pred_svm <- c(pred_svm, result_svm$pred_class)
    result_knn <- train_and_evaluate(train_data, test_data, "knn") # Evaluate KNN
    pred_knn <- c(pred_knn, result_knn$pred_class)
    result_nb <- train_and_evaluate(train_data, test_data, "nb") # Evaluate Naive Bayes
    pred_nb <- c(pred_nb, result_nb$pred_class)
    real_classes <- c(real_classes, test_data$AMI) # Store the actual class
  }
  
  # Calculate AUC for each model
  auc_bn <- roc(real_classes, as.numeric(pred_bn))$auc
  auc_rf <- roc(real_classes, as.numeric(pred_rf))$auc
  auc_svm <- roc(real_classes, as.numeric(pred_svm))$auc
  auc_knn <- roc(real_classes, as.numeric(pred_knn))$auc
  auc_nb <- roc(real_classes, as.numeric(pred_nb))$auc
  # Print AUC for each model
  cat("AUC Bayesian Network: ",auc_bn, "\n")
  cat("AUC Random Forest: ",auc_rf, "\n")
  cat("AUC SVM: ",auc_svm, "\n")
  cat("AUC KNN: ",auc_knn, "\n")
  cat("AUC Naive Bayes: ", auc_nb, "\n")
  # store the results in the data frame
  auc_results <- rbind(auc_results, data.frame(AUC = auc_bn, Model = "Bayesian Network", Iteration = iter))
  auc_results <- rbind(auc_results, data.frame(AUC = auc_rf, Model = "Random Forest", Iteration = iter))
  auc_results <- rbind(auc_results, data.frame(AUC = auc_svm, Model = "SVM", Iteration = iter))
  auc_results <- rbind(auc_results, data.frame(AUC = auc_knn, Model = "KNN", Iteration = iter))
  auc_results <- rbind(auc_results, data.frame(AUC = auc_nb, Model = "Naive Bayes", Iteration = iter))
  # store specificity and sensitivity for each model
  specificity_list[["Bayesian Network"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_bn))$specificities)
  sensitivity_list[["Bayesian Network"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_bn))$sensitivities)
  specificity_list[["Random Forest"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_rf))$specificities)
  sensitivity_list[["Random Forest"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_rf))$sensitivities)
  specificity_list[["SVM"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_svm))$specificities)
  sensitivity_list[["SVM"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_svm))$sensitivities)
  specificity_list[["KNN"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_knn))$specificities)
  sensitivity_list[["KNN"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_knn))$sensitivities)
  specificity_list[["Naive Bayes"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_nb))$specificities)
  sensitivity_list[["Naive Bayes"]][[iter]] <- rev(roc(real_classes, as.numeric(pred_nb))$sensitivities)
}
# Calculate the mean specificity and sensitivity for each model
mean_specificity <- lapply(specificity_list, function(x) rowMeans(do.call(cbind, x), na.rm = TRUE))
mean_sensitivity <- lapply(sensitivity_list, function(x) rowMeans(do.call(cbind, x), na.rm = TRUE))
# Convert mean ROC values to data frames
roc_curves_mean_df <- lapply(names(mean_specificity), function(model_name) {
  data.frame(
    specificity = mean_specificity[[model_name]],
    sensitivity = mean_sensitivity[[model_name]],
    Model = model_name
  )
})
# Combine all mean ROC data frames into one
roc_curves_mean_combined <- do.call(rbind, roc_curves_mean_df)
# Plot the AUC results
ggplot(auc_results, aes(x = Model, y = AUC))+geom_boxplot()+ggtitle("AUC Distribution for Each Model (10 iterations of undersampling)")+ylab("AUC")+xlab("Model")

# Plot mean ROC curves
ggplot(roc_curves_mean_combined, aes(x = 1 - specificity, y = sensitivity, color = Model, linetype = Model)) +geom_line(linewidth = 1)+ggtitle("Mean ROC Curves for Different Models (10 Iterations)")+xlab("1 - Specificity")+ylab("Sensitivity")+theme_minimal()+scale_color_manual(values = c("red", "blue", "green", "purple", "orange"), labels = names(mean_specificity))
```
