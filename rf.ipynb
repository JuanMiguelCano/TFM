```r
library(readxl)
library(randomForest)
```

```r
# antecedentes
data <- read_excel("antecedentes.xlsx")
n <- nrow(data)
set.seed(1)
indtrain <- sample(1:n, round(0.7*n))
indtest <- setdiff(1:n, indtrain)

predictores <- subset(data, select = -c(MuerteCV, Revasc))
data$MuerteCV <- as.factor(data$MuerteCV)
data$Revasc <- as.factor(data$Revasc)
rfmuertecv <- randomForest(x = predictores[indtrain,], y = data$MuerteCV[indtrain])
rfrevasc <- randomForest(x = predictores[indtrain,], y = data$Revasc[indtrain])
# print(rfmuertecv$ntree)
# print(rfrevasc$ntree)
varImpPlot(rfmuertecv)
varImpPlot(rfrevasc)
```
